```python
import json
import re
```

```python
def readTextFiles(fileName):
    """
    Read the text files to get json data
    """
    if(fileName == None):
        print("Missing File Name")
        return

    for line in open(fileName, mode="r"):
        jstring = json.loads(line)   
    nodeList = jstring['response']
    
    """
    get timestamp of eachFile
    var name stores the timestamp in integer value
    """
    name = re.split(r'[_.]',fileName)[1]
    articleID = re.split(r'[_.]',fileName)[0]
    
    for node in nodeList:
        node['timeStamp'] = int(name)
        
    return nodeList, articleID


def writeToFile(nodes, fileName):
    """
    Write the node to a new file
    """
    if fileName==None:
        print("Missing FileName")
        return
    
    jsonData = json.dumps(nodes) 
    with open(fileName, "w") as file:
        #write the node to the file
        file.write(jsonData)
        file.write(",")
        
    print("Output written to file: " + fileName)
    return


def findDuplicates(node1, node2):
    """
    Check if two nodes have the same id
    each node is a dict
    """
    #Check if node ids match.
    if(node1 == None):
        return false
    if node1['id'] == node2['id']:
        #print("Duplicates")
        return True
    else:
        return False

def main():
    fileNames = []
    """
    Infinte loop to read all file names until user says submit
    """
    while True:
        fileName = str(input("Enter each file name. Enter 'Submit' to start merging: "))
        if(fileName == 'Submit'):
            break
        fileNames.append(fileName)
    
    #Sort the file names
    fileNames.sort()
    
    """
    listAllNodes is the final list containing all json comment nodes
    """
    listAllNodes = [] #stores all combined nodes
    
    print("Merging files...")
    
    for index, file in enumerate(fileNames,1):
        list1, articleID = readTextFiles(file)
        newFileName = articleID+ ".txt"
            
        for node in list1:
            dup = 0 #flag for duplicates
            for dictionaryNode in listAllNodes:
                #print("Checking for duplicates. Calling duplicates method")
                duplicates = findDuplicates(node, dictionaryNode) 
                if(duplicates == True):
                    #check for latest node
                    dup = 1
                    if(node['timeStamp'] > dictionaryNode['timeStamp']):
                        #write the update timestamp
                        listAllNodes.append(node)
                        listAllNodes.remove(dictionaryNode)
                        break
            if(dup == 0):
                #write the node
                listAllNodes.append(node)
        
        #print("Size of ListNode This Iteration: "+ str(len(listAllNodes)))
        if(index % 3 == 0):
            #print("Size of List: "+ str(len(listAllNodes)))
            #print("Output File Name: "+ newFileName)
            writeToFile(listAllNodes, newFileName)
            listAllNodes.clear()
    #writeToFile(listAllNodes, newFileName)
```
